```python
# Verificar vers√µes instaladas
import pyspark
print(f"PySpark: {pyspark.__version__}")

try:
    import delta
    print("Delta Lake: OK")
except Exception as e:
    print(f"Delta Lake: ERRO - {e}")

# Iniciar Spark
from pyspark.sql import SparkSession
from delta import configure_spark_with_delta_pip

builder = SparkSession.builder \
    .appName("Teste") \
    .master("local[*]") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog")

spark = configure_spark_with_delta_pip(builder).getOrCreate()
spark.sparkContext.setLogLevel("ERROR")

print(f"\n‚úÖ Spark {spark.version} OK!")
```

```python

# Teste simples SEM Delta primeiro
print("\nüìù Teste simples (Parquet)...")
df = spark.createDataFrame([(1, "Jo√£o"), (2, "Maria")], ["id", "nome"])
df.write.mode("overwrite").parquet("D:\Projetos\Jornada_financas_pessoais\parquet")
print("‚úÖ Parquet OK")

# Agora teste Delta
print("\nüìù Teste Delta Lake...")
df.write.format("delta").mode("overwrite").save("D:\Projetos\Jornada_financas_pessoais\delta")
print("‚úÖ Delta Lake OK!")

spark.stop()
```
